```python
import pandas as pd
from pathlib import Path

DATA_DIR = Path("../data")

files = {
    "local": DATA_DIR / "local_data.xlsx",
    "nigeria": DATA_DIR / "CompiledjobListNigeria.csv",
    "fake_postings": DATA_DIR / "Fake Postings.csv",
    "kaggle": DATA_DIR / "fake_job_postings.csv"
}

local_df = pd.read_excel(files["local"])
nigeria_df = pd.read_csv(files["nigeria"])
fake_postings_df = pd.read_csv(files["fake_postings"])
kaggle_df = pd.read_csv(files["kaggle"])

datasets = {
    "local": local_df,
    "nigeria": nigeria_df,
    "fake_postings": fake_postings_df,
    "kaggle": kaggle_df
}

for name, df in datasets.items():
    print(f"\n===== {name.upper()} DATASET =====")
    print("Shape:", df.shape)
    print("Columns:", df.columns.tolist())
    display(df.head())
```
